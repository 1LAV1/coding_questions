```python
def find_all_occurrences(main_string, substring):
    occurrences = []
    start = 0
    while True:
        start = main_string.find(substring,start)
        if start == -1:
            break
        occurrences.append(start)
        start += 1
    return occurrences


main_string = input("enter main string")#helloworldhello
substring = input("enter substring")#hello
print(find_all_occurrences(main_string, substring)) #[0,10]

```

```python
#  write code to perform basic string compression by counting the repeated characters.

#code
def compress_string(string):
    compressed_string = ""
    count = 1

    for i in range(len(string)):
        if i < len(string) - 1 and string[i] == string[i + 1]:
            count += 1
        else:
            compressed_string += string[i] + str(count)
            count = 1

    return compressed_string if len(compressed_string) < len(string) else string


original_string = input("Enter the original string: ")
compressed_result = compress_string(original_string)
print("Compressed string:", compressed_result)
```

```python
# write code to  to determine if a string has all unique characters.
#code
def has_unique_characters(string):
    char_set = set()
    for char in string:
        if char in char_set:
            return False
        char_set.add(char)
    return True

input_string = input("enter the string")
if has_unique_characters(input_string):
    print("The string has all unique characters.")
else:
    print("The string does not have all unique characters.")
```

```python
# write code to find the second largest number in a given list of integers
#code
def second_largest(lst):
    if len(lst) < 2:
        return None

    largest = max(lst[0], lst[1])
    second_largest = min(lst[0], lst[1])

    for num in lst[2:]:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest


nums = list(map(int, input().split()))
result = second_largest(nums)
print("Second largest number:", result)
```

```python
#code
def reverse_list(lst):
    left = 0
    right = len(lst) - 1

    while left < right:

        lst[left], lst[right] = lst[right], lst[left]

        left += 1
        right -= 1


my_list = list(map(int, input().split()))
reverse_list(my_list)
print(my_list)
```

```python
# write code to  implement it to find and remove duplicates from a list while preserving the original order



#code
def remove_duplicates(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result


original_list =  list(map(int, input().split()))
unique_list = remove_duplicates(original_list)
print(unique_list)
```

```python
#write code if a given list is sorted (either in ascending or descending order) by comparing adjacent elements

#code
def is_sorted(lst):

    if all(lst[i] <= lst[i + 1] for i in range(len(lst) - 1)):
        return "Ascending"


    if all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1)):
        return "Descending"


    return "Unsorted"


l = list(map(int, input().split()))
print(is_sorted(l))



```

```python

#write code to shuffle a list randomly without using any built-in shuffle function


#code
import random

def shuffle_list(lst):
    for i in range(len(lst) - 1, 0, -1):
        j = random.randint(0, i)
        lst[i], lst[j] = lst[j], lst[i]


my_list = [1, 2, 3, 4, 5]
shuffle_list(my_list)
print("Shuffled List:", my_list)
```

```python
# write  code that takes two dictionaries as input and merges them into a single dictionary. If they share common keys, the values should be added together





#code
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value
    return merged_dict


dict1 = {'apple': 3, 'banana': 2, 'orange': 1}
dict2 = {'apple': 2, 'banana': 1, 'grape': 4}
merged_dict = merge_dictionaries(dict1, dict2)
print("Merged Dictionary:", merged_dict)
```

```python
#write  code that accesses the values in a nested dictionary. The function takes a dictionary and a list of keys as input and returns the corresponding values




#code
def access_nested_dict(nested_dict, keys_list):
    result = nested_dict
    for key in keys_list:
        if key in result:
            result = result[key]
        else:
            return None
    return result

# Example usage:
nested_dictionary = {
    'fruit': {
        'apple': 3,
        'banana': 2,
        'orange': 1
    },
    'color': {
        'red': 5,
        'blue': 4,
        'green': 3
    }
}
keys = ['fruit', 'apple']
value = access_nested_dict(nested_dictionary, keys)
print("Value corresponding to keys:", value)
```

```python
#write code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order




#code
def sort_dictionary_by_values(input_dict, ascending=True):
    sorted_dict = dict(sorted(input_dict.items(), key=lambda x: x[1], reverse=not ascending))
    return sorted_dict


input_dict = {'apple': 3, 'banana': 2, 'orange': 1}
sorted_dict_asc = sort_dictionary_by_values(input_dict, ascending=True)
sorted_dict_desc = sort_dictionary_by_values(input_dict, ascending=False)

print("Sorted Dictionary (Ascending):", sorted_dict_asc)
print("Sorted Dictionary (Descending):", sorted_dict_desc)
```

```python
#write code that inverts a dictionary by swapping keys and values.




#code
def invert_dictionary(input_dict):
    inverted_dict = {}
    for key, value in input_dict.items():
        if value not in inverted_dict:
            inverted_dict[value] = key
        else:
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
    return inverted_dict

input_dict = {'apple': 3, 'banana': 2, 'orange': 1, 'grape': 3}
inverted_dict = invert_dictionary(input_dict)
print("Inverted Dictionary:", inverted_dict)
```
